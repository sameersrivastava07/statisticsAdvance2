Question:1--->> What is hypothesis testing in statistics?

Answer:1-->>>

1. Hypothesis testing is a method in statistics that helps us make decisions or judgments about a population based on a sample.

2. We usually begin with a statement called the **null hypothesis (H₀)** which assumes that there is no effect or no difference. For example, we might say "There is no change in average marks after coaching."

3. Along with that, we have an **alternative hypothesis (H₁ or Ha)** which says that there *is* an effect or a difference. Like, "There is a change in average marks after coaching."

4. Then we collect data (like test scores) and perform a statistical test (like Z-test, t-test, etc.) to check whether the sample data supports the null hypothesis or not.

5. If the result is unlikely under the null hypothesis (usually based on a small p-value), we reject the null hypothesis and accept the alternative.

6. For example, Sameer wanted to know if a new diet plan worked better for his weight loss. He used hypothesis testing to check if the average weight after following the plan was significantly different from before.

7. So, hypothesis testing helps us make data-based conclusions instead of guessing or assuming things.


Question:2--->> What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer:2-->>>

1. The null hypothesis (H₀) is a statement that assumes there is no effect, no difference, or no relationship between variables. It is the starting point in hypothesis testing.

2. It is considered true until we find strong evidence to reject it through statistical testing.

3. The alternative hypothesis (H₁ or Ha) is the statement that suggests there is an effect, a difference, or a relationship between variables.

4. Basically, the null hypothesis says "nothing is happening", while the alternative hypothesis says "something is happening".

5. For example, Sameer wants to test if a new coaching class improved student performance.

   * Null hypothesis (H₀): The coaching class has no effect on students' marks.
   * Alternative hypothesis (H₁): The coaching class improved the students' marks.

6. So, the difference is that the null hypothesis assumes no change or impact, and the alternative hypothesis assumes there is a change or impact.


Question:3--->> What is the significance level in hypothesis testing, and why is it important?

Answer:3-->>>

1. The significance level in hypothesis testing is a value that shows how much risk we are willing to take of rejecting the null hypothesis when it is actually true.

2. It is usually written as alpha (α), and common values are 0.05, 0.01, or 0.10.

3. For example, if α = 0.05, it means we are ready to accept a 5% chance of making a wrong decision by rejecting a true null hypothesis.

4. The significance level helps us decide whether our test result is strong enough to reject the null hypothesis. If the p-value is less than α, we reject the null.

5. It is important because it controls the probability of making a **Type I error**, which means saying there is an effect when there is none.

6. For example, Shweta tested if a new fertilizer increases plant growth. If her test gives a p-value of 0.03 and she set α = 0.05, she will reject the null and believe the fertilizer works.

7. So, the significance level acts like a cutoff point. It helps us make safe and confident decisions using data instead of guessing.


Question:4--->> What does a P-value represent in hypothesis testing?

Answer:4-->>>

1. A p-value is a number that helps us understand how likely it is to get the results we got, if the null hypothesis is actually true.

2. It tells us the probability of observing the sample results (or something more extreme) just by random chance.

3. A small p-value means that our sample results are unlikely under the null hypothesis, so we may reject the null.

4. A large p-value means the results are likely under the null hypothesis, so we don’t have strong reason to reject it.

5. For example, if Sameer performs a test and gets a p-value of 0.02, and he set the significance level at 0.05, then 0.02 < 0.05. So he will reject the null hypothesis.

6. In simple words, the p-value helps us decide whether our data gives strong enough evidence against the null hypothesis.

7. So, the smaller the p-value, the stronger the evidence that we should reject the null hypothesis.


Question:5--->> How do you interpret the P-value in hypothesis testing?

Answer:5-->>>

1. The p-value tells us the probability of getting the observed result, or something more extreme, if the null hypothesis is actually true.

2. If the p-value is less than or equal to the significance level (for example, 0.05), we reject the null hypothesis. This means there is strong evidence in favour of the alternative hypothesis.

3. If the p-value is greater than the significance level, we fail to reject the null hypothesis. This means the data does not provide strong enough evidence against it.

4. For example, Sameer tested if a new app increased user engagement and got a p-value of 0.03. Since 0.03 is less than 0.05, he rejects the null hypothesis and believes the app made a difference.

5. But if Shweta tested something and got a p-value of 0.10, she would not reject the null because 0.10 is greater than 0.05. So, she doesn't have strong proof that anything changed.

6. In simple words, the smaller the p-value, the stronger the evidence against the null hypothesis.


Question:6--->> What are Type 1 and Type 2 errors in hypothesis testing?

Answer:6-->>>

1. Type 1 and Type 2 errors are two possible mistakes that can happen when we make decisions in hypothesis testing.

2. A Type 1 error happens when we reject the null hypothesis even though it is actually true. It means we think something has changed, but in reality, nothing changed.

3. For example, Sameer tested if a new online course improves marks. If the course actually has no effect, but he still says it does, that is a Type 1 error.

4. A Type 2 error happens when we fail to reject the null hypothesis even though it is actually false. It means we miss a real change.

5. For example, Shweta tested a new diet plan. It actually works, but the test results didn’t show enough proof. So she accepts the null and says it doesn’t work. That is a Type 2 error.

6. Type 1 error is linked to the significance level (alpha), while Type 2 error is related to the power of the test.

7. So, Type 1 error means a false alarm, and Type 2 error means a missed opportunity.


Question:7--->> What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

Answer:7-->>>

1. A one-tailed test is used when we want to check if a value is either **greater than** or **less than** a certain number, but not both.

2. It focuses on only one side of the distribution curve, so the full significance level (like 0.05) is applied to one direction.

3. For example, Sameer wants to test if the new training program increases sales. He will use a one-tailed test because he is only interested in whether sales went up.

4. A two-tailed test is used when we want to check if a value is **different** from a certain number, in **either direction**—greater or less.

5. In a two-tailed test, the significance level is split into two parts (like 0.025 on each side if alpha is 0.05).

6. For example, Shweta wants to test if students’ average marks are not equal to 60. She will use a two-tailed test because marks could be either higher or lower.

7. So, the main difference is that a one-tailed test checks in one direction, while a two-tailed test checks in both directions.


Question:8--->> What is the Z-test, and when is it used in hypothesis testing?

Answer:8-->>>

1. A Z-test is a type of statistical test used to check if there is a significant difference between sample data and population data, or between two sample means.

2. It is used when the population variance is known and the sample size is large (usually more than 30).

3. The Z-test helps in testing hypotheses about population means and proportions.

4. For example, Sameer wants to test if the average height of students in his college is different from the national average. If he knows the population standard deviation, he can use a Z-test.

5. The Z-test calculates a Z-score, which tells how many standard deviations the sample mean is away from the population mean.

6. If the Z-score leads to a p-value smaller than the chosen significance level, we reject the null hypothesis.

7. So, the Z-test is mainly used for large samples when the standard deviation of the population is known, and we want to compare means or proportions.


Question:9--->> How do you calculate the Z-score, and what does it represent in hypothesis testing?

Answer:9-->>>

1. The Z-score is calculated using the formula:
   Z = (X̄ − μ) / (σ / √n)
   where X̄ is the sample mean, μ is the population mean, σ is the population standard deviation, and n is the sample size.

2. The Z-score tells us how many standard deviations the sample mean is away from the population mean.

3. In hypothesis testing, the Z-score helps us decide whether the sample result is far enough from the null hypothesis to reject it.

4. For example, Shweta collected data on student marks and got a Z-score of 2.1. This means her sample mean is 2.1 standard deviations above the population mean.

5. After calculating the Z-score, we compare it to a critical value (based on the significance level). If it falls in the rejection region, we reject the null hypothesis.

6. So, the Z-score represents how extreme or unusual the sample result is compared to what we expect under the null hypothesis.


Question:10--->> What is the T-distribution, and when should it be used instead of the normal distribution?

Answer:10-->>>

1. The T-distribution is a type of probability distribution that is similar to the normal distribution but has fatter tails.

2. It is used in hypothesis testing when the sample size is small (usually less than 30) and the population standard deviation is not known.

3. Unlike the normal distribution, the T-distribution changes shape based on the sample size. It becomes more like the normal curve as the sample size increases.

4. For example, Sameer collected data from only 15 students to test average study hours. Since the sample is small and population standard deviation is unknown, he should use the T-distribution.

5. The T-distribution gives more accurate results in small samples because it accounts for the extra uncertainty in estimating the standard deviation from the sample.

6. So, we use the T-distribution instead of the normal distribution when the sample size is small and the population standard deviation is unknown.


Question:11--->> What is the difference between a Z-test and a T-test?

Answer:11-->>>

1. A Z-test is used when the sample size is large (usually n > 30) and the population standard deviation is known.

2. A T-test is used when the sample size is small (usually n ≤ 30) and the population standard deviation is unknown.

3. The Z-test follows the normal distribution, while the T-test follows the T-distribution which has heavier tails to handle more variability in small samples.

4. For example, Sameer surveyed 100 students and knew the population standard deviation. So he used a Z-test.

5. But when Shweta surveyed only 20 students and didn’t know the population standard deviation, she used a T-test.

6. So, the main difference is that Z-test is for large samples with known population standard deviation, and T-test is for small samples with unknown population standard deviation.


Question:12--->> What is the T-test, and how is it used in hypothesis testing?

Answer:12-->>>

1. A T-test is a statistical test used to compare the mean of a sample to a known value or to compare the means of two groups.

2. It is mainly used when the sample size is small (usually less than or equal to 30) and the population standard deviation is unknown.

3. The T-test uses the T-distribution to calculate a T-score, which tells how far the sample mean is from the population mean in terms of standard error.

4. For example, Sameer wants to test if the average sleep time of college students is different from 7 hours. Since he only surveyed 20 students and doesn't know the population standard deviation, he will use a T-test.

5. After calculating the T-score, we compare it with the critical value from the T-distribution table based on degrees of freedom and significance level.

6. If the T-score falls in the rejection region, we reject the null hypothesis, meaning there is a significant difference.

7. So, the T-test helps us decide if the observed difference in means is real or just due to random chance in small samples.


Question:13--->> What is the relationship between Z-test and T-test in hypothesis testing?

Answer:13-->>>

1. Both Z-test and T-test are used in hypothesis testing to compare means and check if the observed results are significantly different from the expected results.

2. They follow similar logic: both calculate a test statistic (Z-score or T-score) and compare it to a critical value to make a decision about the null hypothesis.

3. The main difference is in the conditions where they are used.

   * Z-test is used when the sample size is large and population standard deviation is known.
   * T-test is used when the sample size is small and population standard deviation is unknown.

4. As the sample size increases, the T-distribution becomes similar to the normal distribution. So in large samples, the T-test and Z-test give almost the same results.

5. For example, Sameer used a Z-test for comparing heights in a large group where the standard deviation was known. Shweta used a T-test when she only had data from a small group and didn’t know the standard deviation.

6. So, the relationship is that both tests are used for similar purposes, but the choice between them depends on sample size and whether the population standard deviation is known.


Question:14--->> What is a confidence interval, and how is it used to interpret statistical results?

Answer:14-->>>

1. A confidence interval is a range of values that is used to estimate the true value of a population parameter, like the mean or proportion.

2. It gives us an idea of how uncertain or accurate our sample estimate is.

3. A common confidence level is 95%, which means we are 95% confident that the true population value lies within the interval.

4. For example, Sameer calculated the average marks of students and got a confidence interval of 68 to 74. This means he is 95% confident that the true average marks of all students lie between 68 and 74.

5. The confidence interval is calculated using the sample mean, standard error, and the critical value (Z or T based on sample size).

6. If the confidence interval includes the value from the null hypothesis, we usually do not reject the null. If it doesn’t include that value, we may reject the null hypothesis.

7. So, confidence intervals help us understand the reliability of our sample results and make better decisions based on data.


Question:15--->> What is the margin of error, and how does it affect the confidence interval?

Answer:15-->>>

1. The margin of error is the amount added and subtracted from the sample estimate to create a confidence interval.

2. It shows how much uncertainty there is in our estimate due to sampling.

3. The formula for margin of error is:
   Margin of Error = Critical value × Standard error

4. A larger margin of error means the confidence interval will be wider, showing more uncertainty. A smaller margin of error means a narrower interval, showing more precision.

5. For example, Shweta calculated an average height of 160 cm with a margin of error of 5 cm. So, the confidence interval is 155 to 165 cm.

6. Factors that affect the margin of error include the sample size (larger samples give smaller margins) and the confidence level (higher confidence gives larger margins).

7. So, the margin of error directly affects the width of the confidence interval and helps us understand how reliable our estimate is.


Question:16--->> How is Bayes' Theorem used in statistics, and what is its significance?

Answer:16-->>>

1. Bayes' Theorem is a formula used in statistics to find the probability of an event based on prior knowledge or earlier data.

2. It helps us update our beliefs about an event when new evidence or information is available.

3. The formula is:
   P(A|B) = \[P(B|A) × P(A)] / P(B)
   where P(A|B) is the probability of A given B has occurred.

4. For example, Sameer wants to know the chance he has a disease after testing positive. Bayes’ Theorem helps him calculate this by using the test’s accuracy and the general rate of the disease.

5. It is useful in fields like medicine, spam filtering, machine learning, and decision-making under uncertainty.

6. The significance of Bayes’ Theorem is that it allows us to improve predictions and decisions by combining past knowledge with new data.

7. So, Bayes’ Theorem is important because it connects previous beliefs with fresh evidence to give more accurate and updated results.


Question:17--->> What is the Chi-square distribution, and when is it used?

Answer:17-->>>

1. The Chi-square distribution is a type of probability distribution used mainly for testing relationships between categorical variables.

2. It is not symmetric like the normal distribution and is always positive since it is based on squared values.

3. The Chi-square test is commonly used in two situations:
   a) To test **goodness of fit** – whether the observed data matches expected data.
   b) To test **independence** – whether two categorical variables are related or not.

4. For example, Sameer wants to check if the preference for different mobile brands is the same among college students. He can use a Chi-square test.

5. The test compares observed frequencies with expected frequencies. If the difference is big enough, we may reject the null hypothesis.

6. The shape of the Chi-square distribution depends on the degrees of freedom. As the degrees of freedom increase, it becomes more like a normal curve.

7. So, the Chi-square distribution is used when we deal with categorical data and want to check patterns, associations, or how well data fits expected outcomes.


Question:18--->> What is the Chi-square goodness of fit test, and how is it applied?

Answer:18-->>>

1. The Chi-square goodness of fit test is used to check whether the observed frequencies in a category match the expected frequencies.

2. It helps us decide if a sample data set fits a particular theoretical distribution or pattern.

3. The formula used is:
   χ² = Σ \[(O − E)² / E]
   where O is the observed frequency and E is the expected frequency.

4. For example, Sameer rolls a die 60 times and gets different numbers of 1s, 2s, etc. He wants to check if the die is fair. He will use the Chi-square goodness of fit test.

5. He compares the observed frequencies of each number with the expected frequency (which would be 10 for each face in a fair die).

6. After calculating the test statistic, he compares it with a critical value from the Chi-square distribution table based on degrees of freedom and significance level.

7. If the calculated value is greater than the critical value, he will reject the null hypothesis, meaning the observed data does not fit the expected pattern.

8. So, this test is applied when we want to check how well our observed data fits what we expected to happen.


Question:19--->> What is the F-distribution, and when is it used in hypothesis testing?

Answer:19-->>>

1. The F-distribution is a type of probability distribution used mainly to compare two variances or test overall group differences.

2. It is not symmetric and always has values greater than or equal to zero.

3. The F-distribution is mainly used in the **ANOVA test (Analysis of Variance)** and in comparing the variances of two samples.

4. For example, Sameer wants to check if three different coaching centers give different average scores. He uses ANOVA, which is based on the F-distribution.

5. The F-test compares the variance **between** groups to the variance **within** groups. If the between-group variance is much higher, we may reject the null hypothesis.

6. The shape of the F-distribution depends on two different degrees of freedom: one for the numerator and one for the denominator.

7. So, the F-distribution is used in hypothesis testing when we want to compare multiple means or variances, especially in ANOVA or regression model testing.


Question:20--->> What is an ANOVA test, and what are its assumptions?

Answer:20-->>>

1. ANOVA stands for **Analysis of Variance**, and it is a statistical test used to compare the means of **three or more groups** to see if at least one group is significantly different from the others.

2. It uses the **F-distribution** to test if the variation between group means is larger than the variation within the groups.

3. For example, Sameer wants to check if students from three different schools have different average maths scores. ANOVA can help him compare all three at once instead of doing multiple t-tests.

4. The **null hypothesis** in ANOVA says all group means are equal, and the **alternative hypothesis** says at least one mean is different.

5. The main **assumptions** of ANOVA are:
   a) The populations from which the samples are taken are **normally distributed**.
   b) All groups have **equal variances** (homogeneity of variance).
   c) The samples are **independent** from each other.
   d) The data should be **measured at interval or ratio level** (not just categories).

6. If the ANOVA result is significant, we reject the null hypothesis, but we don’t know which group is different. In that case, we do further tests like **post-hoc tests**.

7. So, ANOVA is useful when comparing multiple groups at once, and it gives reliable results only if its basic assumptions are met.


Question:21--->> What are the different types of ANOVA tests?

Answer:21-->>>

1. There are mainly three types of ANOVA tests, and each is used based on how many factors or groups we are comparing.

2. One-Way ANOVA:
   This test is used when we compare the means of three or more groups based on **one independent variable**.
   For example, Sameer compares the average marks of students from three different schools using one-way ANOVA.

3. Two-Way ANOVA:
   This test is used when we study the effect of **two independent variables** on a dependent variable.
   For example, Shweta checks how both study method (online/offline) and gender affect exam performance.

4. Repeated Measures ANOVA:
   This test is used when the **same group** is measured multiple times under different conditions or time periods.
   For example, Sameer measures the performance of students before coaching, after 1 month, and after 2 months.

5. The choice of ANOVA test depends on how many factors are being studied and whether the data is from the same group or different groups.

6. So, ANOVA has different types to handle various testing situations depending on the design of the experiment or study.


Question:22--->> What is the F-test, and how does it relate to hypothesis testing?

Answer:22-->>>

1. The F-test is a statistical test used to compare two variances or to check if multiple group means are significantly different.

2. It is mainly used in **ANOVA** and in comparing the fit of two regression models or checking the equality of variances.

3. The F-test works by calculating an **F-value**, which is the ratio of two variances. If the F-value is large enough, it suggests that there is a significant difference.

4. For example, Sameer wants to compare the exam scores of students from three coaching centers. He uses an F-test through ANOVA to check if the average scores are different.

5. The null hypothesis in an F-test usually says that all variances or group means are equal, and the alternative says at least one is different.

6. If the calculated F-value is greater than the critical value from the F-distribution table, we reject the null hypothesis.

7. So, the F-test is closely related to hypothesis testing as it helps us decide whether the differences we observe are real or just due to chance.

```python
# # Question:5--->> Create a Poisson distribution and visualize it using Python?

# Answer:5-->>>

# Import necessary libraries
import numpy as np                    # For random Poisson values
import matplotlib.pyplot as plt       # For plotting
from scipy.stats import poisson       # For Poisson PMF

# Step 1: Define the Poisson distribution parameter (λ)
# Let's say Sameer records average 4 calls per hour at a call center
lam = 4

# Step 2: Generate random data from Poisson distribution
data = np.random.poisson(lam=lam, size=1000)  # Generate 1000 random values

# Step 3: Count frequency of each unique value
unique_values, counts = np.unique(data, return_counts=True)

# Step 4: Calculate the PMF for plotting
x = np.arange(0, max(data)+1)
pmf = poisson.pmf(x, mu=lam)

# Step 5: Plot histogram of the simulated data
plt.figure(figsize=(8, 4))
plt.bar(unique_values, counts/1000, width=0.6, alpha=0.6, label="Simulated Data")
plt.plot(x, pmf, 'ro-', label="Poisson PMF (λ=4)")
plt.title("Poisson Distribution (λ = 4)")
plt.xlabel("Number of events (k)")
plt.ylabel("Probability")
plt.xticks(x)
plt.legend()
plt.grid(True)
plt.show()

# This plot shows both the histogram of simulated values and the theoretical Poisson PMF curve.

```

```python
# Question:6--->> Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution?

# Answer:6-->>>

# Import necessary libraries
import numpy as np                    # For numerical operations
import matplotlib.pyplot as plt       # For plotting
from scipy.stats import randint       # For discrete uniform distribution

# Define the range for the discrete uniform distribution
# Let's say Sameer is choosing a number randomly between 1 and 6 (like a dice)
low = 1
high = 7  # Upper limit is exclusive in randint

# Create the distribution object
dist = randint(low, high)

# Create the list of values for x-axis (possible outcomes)
x = np.arange(low, high)

# Calculate CDF values for each x
cdf_values = dist.cdf(x)

# Plot the CDF
plt.step(x, cdf_values, where='post')  # step grap_
```

```python
# Question:10--->> Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution?

# Answer:10-->>>

# Import necessary libraries
import numpy as np                  # For data generation and calculations
import matplotlib.pyplot as plt     # For plotting the histogram

# Step 1: Generate a non-normal (skewed) population
# Let's say Shweta is collecting data of waiting times, which are usually right-skewed
population = np.random.exponential(scale=2.0, size=100000)

# Step 2: Create a function to apply CLT by taking multiple samples
def apply_clt(population_data, sample_size=30, num_samples=1000):
    sample_means = []

    # Draw 'num_samples' random samples from the population
    for _ in range(num_samples):
        # Randomly select 'sample_size' elements from the population
        sample = np.random.choice(population_data, size=sample_size)

        # Calculate the mean of the sample
        sample_mean = np.mean(sample)

        # Store the mean
        sample_means.append(sample_mean)

    # Plot the histogram of the sample means
    plt.hist(sample_means, bins=30, color='green', edgecolor='black', density=True)
    plt.xlabel('Sample Means')
    plt.ylabel('Density')
    plt.title(f'Central Limit Theorem - Sample Size = {sample_size}')
    plt.grid(True)
    plt.show()

# Apply CLT function using the skewed exponential population
apply_clt(population, sample_size=30, num_samples=1000)

#This will show a bell-shaped histogram of sample means even though the original data was skewed
```

```python
# Question:15--->> Implement hypothesis testing using Z-statistics for a sample dataset?

# Answer:15-->>>

# Import necessary libraries
import numpy as np                  # For calculations
from scipy.stats import norm        # For Z critical value and p-value

# Step 1: Define the sample dataset
# Let's say Sameer collected test scores of 50 students
sample_data = np.array([68, 72, 75, 70, 69, 73, 74, 71, 70, 72,
                        68, 69, 70, 71, 74, 73, 72, 70, 71, 75,
                        73, 72, 69, 70, 74, 72, 73, 71, 70, 68,
                        69, 70, 71, 72, 73, 74, 75, 70, 69, 70,
                        71, 72, 73, 74, 75, 70, 71, 72, 73, 74])

# Step 2: Set known population parameters
population_mean = 70          # Known population mean
population_std = 3            # Known population standard deviation
n = len(sample_data)          # Sample size
sample_mean = np.mean(sample_data)  # Sample mean

# Step 3: Calculate Z-statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Step 4: Decide significance level (alpha)
alpha = 0.05

# Step 5: Calculate critical Z value for two-tailed test
z_critical = norm.ppf(1 - alpha/2)

# Step 6: Print results
print("Sample Mean:", sample_mean)
print("Z-statistic:", z_stat)
print("Z-critical value (two-tailed): ±", z_critical)

# Step 7: Make decision
if abs(z_stat) > z_critical:
    print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")

#This will show whether Sameer's sample mean is significantly different from the population mean
```

```python
# Question:16--->> Create a confidence interval for a dataset using Python and interpret the result?

# Answer:16-->>>

# Import necessary libraries
import numpy as np                  # For numerical operations
from scipy.stats import norm        # For Z-value

# Step 1: Define a sample dataset
# Let's say Shweta collected marks of 30 students
data = np.array([68, 70, 72, 65, 74, 75, 73, 69, 71, 70,
                 68, 67, 69, 72, 73, 75, 70, 71, 69, 68,
                 72, 73, 74, 70, 71, 69, 67, 66, 68, 70])

# Step 2: Calculate sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
n = len(data)

# Step 3: Set confidence level (e.g., 95%)
confidence_level = 0.95
alpha = 1 - confidence_level
z_critical = norm.ppf(1 - alpha/2)  # Z-value for 95% confidence

# Step 4: Calculate margin of error
margin_of_error = z_critical * (sample_std / np.sqrt(n))

# Step 5: Calculate confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Step 6: Print results
print(f"Sample Mean: {sample_mean}")
print(f"Confidence Interval (95%): ({lower_bound}, {upper_bound})")

# Interpretation:
# We are 95% confident that the true average marks of the population lie between the lower and upper bounds
```

```python
# Question:1--->> Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results?

# Answer:1-->>>

# Import necessary libraries
import numpy as np                  # For numerical calculations
from scipy.stats import norm        # For Z critical value and p-value

# Step 1: Define the sample data
# Let's say Sameer collected scores of 40 students
sample_data = np.array([78, 82, 85, 80, 79, 81, 84, 83, 80, 78,
                        79, 81, 82, 84, 85, 79, 80, 82, 81, 83,
                        78, 79, 80, 81, 83, 84, 85, 80, 79, 78,
                        82, 83, 84, 80, 79, 81, 82, 83, 79, 80])

# Step 2: Known population parameters
population_mean = 75             # Known population mean
population_std = 5               # Known population standard deviation
n = len(sample_data)             # Sample size

# Step 3: Calculate sample mean
sample_mean = np.mean(sample_data)

# Step 4: Calculate Z-statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Step 5: Define significance level
alpha = 0.05

# Step 6: Calculate critical Z value for two-tailed test
z_critical = norm.ppf(1 - alpha/2)

# Step 7: Calculate p-value for the test
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Step 8: Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"Critical Z value (two-tailed): ±{z_critical:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 9: Make decision based on Z-statistic and p-value
if abs(z_stat) > z_critical:
    print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from population mean.")

# Interpretation:
# If we reject the null, it means the sample provides enough evidence that the average score differs from 75.
```

```python
# Question:4--->> Perform a two-tailed Z-test using Python and visualize the decision region on a plot?

# Answer:4-->>>

# Import necessary libraries
import numpy as np                  # For calculations
import matplotlib.pyplot as plt     # For plotting
from scipy.stats import norm        # For normal distribution functions

# Step 1: Define sample data
# Let's say Sameer has a sample with known population parameters
sample_mean = 105
population_mean = 100
population_std = 15
n = 50  # Sample size

# Step 2: Calculate Z-statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Step 3: Define significance level and critical values for two-tailed test
alpha = 0.05
z_critical_low = norm.ppf(alpha/2)      # Lower critical value (negative)
z_critical_high = norm.ppf(1 - alpha/2) # Upper critical value (positive)

# Step 4: Print calculated Z-statistic and critical values
print(f"Z-statistic: {z_stat:.4f}")
print(f"Critical Z-values: {z_critical_low:.4f}, {z_critical_high:.4f}")

# Step 5: Plot the normal distribution curve and shade rejection regions
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y, label='Standard Normal Distribution')

# Shade rejection regions
plt.fill_between(x, 0, y, where=(x <= z_critical_low), color='red', alpha=0.5, label='Rejection Region (left)')
plt.fill_between(x, 0, y, where=(x >= z_critical_high), color='red', alpha=0.5, label='Rejection Region (right)')

# Mark Z-statistic on the plot
plt.axvline(z_stat, color='blue', linestyle='--', label=f'Z-statistic = {z_stat:.2f}')

# Labels and title
plt.title('Two-tailed Z-test with Decision Regions')
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

# Step 6: Make decision
if z_stat < z_critical_low or z_stat > z_critical_high:
    print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from population mean.")

# Interpretation:
# The plot shows rejection regions in red. If Z-statistic falls in these areas, we reject the null hypothesis.
```

```python
# Question:5--->> Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing?

# Answer:5-->>>

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Define a function to visualize Type 1 and Type 2 errors
def visualize_type1_type2_errors(mu0=100, mu1=105, sigma=15, n=30, alpha=0.05):
    # Calculate standard error
    se = sigma / np.sqrt(n)

    # Calculate critical value (Z) for significance level (alpha)
    z_critical = norm.ppf(1 - alpha)

    # Calculate critical sample mean (cutoff value)
    x_critical = mu0 + z_critical * se

    # Step 2: Create x-axis range for both distributions
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # Step 3: Get normal distributions under H0 and H1
    y_h0 = norm.pdf(x, mu0, se)
    y_h1 = norm.pdf(x, mu1, se)

    # Step 4: Plot both distributions
    plt.plot(x, y_h0, label='H0: µ = 100', color='blue')
    plt.plot(x, y_h1, label='H1: µ = 105', color='green')

    # Step 5: Shade Type 1 Error Region (Reject H0 when H0 is true)
    plt.fill_between(x, 0, y_h0, where=(x >= x_critical), color='red', alpha=0.4, label='Type 1 Error (α)')

    # Step 6: Shade Type 2 Error Region (Fail to Reject H0 when H1 is true)
    plt.fill_between(x, 0, y_h1, where=(x < x_critical), color='orange', alpha=0.4, label='Type 2 Error (β)')

    # Plot vertical line at the critical value
    plt.axvline(x_critical, color='black', linestyle='--', label=f'Critical value = {x_critical:.2f}')

    # Step 7: Add labels, legend, and title
    plt.title('Type 1 and Type 2 Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the function to visualize
visualize_type1_type2_errors()

# Interpretation:
# Red region = Type 1 error (rejecting H0 when it's true)
# Orange region = Type 2 error (failing to reject H0 when H1 is actually true)
```

```python
# Question:6--->> Write a Python program to perform an independent T-test and interpret the results?

# Answer:6-->>>

# Import necessary libraries
import numpy as np                     # For numerical operations
from scipy.stats import ttest_ind      # For independent T-test

# Step 1: Create two sample datasets
# Let's say Sameer and Shweta collected marks from two different batches
batch_A = np.array([78, 80, 82, 79, 77, 81, 80, 83, 78, 79])
batch_B = np.array([74, 76, 75, 73, 77, 75, 74, 72, 76, 75])

# Step 2: Perform an independent T-test (assume equal variance)
t_stat, p_value = ttest_ind(batch_A, batch_B)

# Step 3: Print results
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the two batch means.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the two batch means.")

# Interpretation:
# This test helps us compare the average marks of batch A and B to check if they are statistically different.
```

```python
# Question:8--->> Simulate data and perform both Z-test and T-test, then compare the results using Python?

# Answer:8-->>>

# Import necessary libraries
import numpy as np                      # For simulation
from scipy.stats import ttest_1samp, norm  # For T-test and Z-test
import math                             # For square root

# Step 1: Simulate sample data
# Let's say Shweta simulates test scores of 40 students
np.random.seed(42)
sample_data = np.random.normal(loc=72, scale=10, size=40)

# Step 2: Set known population parameters
population_mean = 70
population_std = 10
n = len(sample_data)

# Step 3: Perform One-sample Z-test
sample_mean = np.mean(sample_data)
z_stat = (sample_mean - population_mean) / (population_std / math.sqrt(n))
p_value_z = 2 * (1 - norm.cdf(abs(z_stat)))

# Step 4: Perform One-sample T-test
t_stat, p_value_t = ttest_1samp(sample_data, population_mean)

# Step 5: Print both results
print(f"Z-test:")
print(f"  Z-statistic = {z_stat:.4f}")
print(f"  P-value = {p_value_z:.4f}")

print(f"\nT-test:")
print(f"  T-statistic = {t_stat:.4f}")
print(f"  P-value = {p_value_t:.4f}")

# Step 6: Interpretation
if p_value_z < 0.05:
    print("\nZ-test: Reject the null hypothesis.")
else:
    print("\nZ-test: Fail to reject the null hypothesis.")

if p_value_t < 0.05:
    print("T-test: Reject the null hypothesis.")
else:
    print("T-test: Fail to reject the null hypothesis.")

# Explanation:
# Z-test is used when population standard deviation is known.
# T-test is used when population standard deviation is unknown.
# Here we see how both tests behave on the same sample data.
```

```python
# Question:9--->> Write a Python function to calculate the confidence interval for a sample mean and explain its significance?

# Answer:9-->>>

# Import necessary libraries
import numpy as np                  # For numerical calculations
from scipy.stats import t           # For critical t value

# Step 1: Define a function to calculate confidence interval
def confidence_interval(data, confidence=0.95):
    # Calculate sample statistics
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)

    # Get the t-critical value for the confidence level
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    # Calculate margin of error
    margin = t_crit * std_err

    # Calculate confidence interval
    lower = mean - margin
    upper = mean + margin

    # Print results
    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")

    # Return the interval
    return (lower, upper)

# Step 2: Use sample data from Sameer's project
sample_scores = np.array([72, 75, 70, 74, 73, 76, 75, 74, 72, 73])

# Step 3: Call the function
confidence_interval(sample_scores)

# Explanation:
# This function calculates the range in which the true population mean is likely to lie.
# The wider the confidence interval, the more uncertainty we have.
# A 95% confidence interval means we're 95% sure the true mean lies in this range.
```

```python
# Question:10--->> Write a Python program to calculate the margin of error for a given confidence level using sample data?

# Answer:10-->>>

# Import necessary libraries
import numpy as np                  # For calculations
from scipy.stats import t           # For t-distribution critical value

# Step 1: Define a function to calculate margin of error
def calculate_margin_of_error(data, confidence=0.95):
    n = len(data)
    sample_mean = np.mean(data)
    std_dev = np.std(data, ddof=1)

    # Calculate standard error
    std_error = std_dev / np.sqrt(n)

    # Find the t-critical value
    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)

    # Calculate margin of error
    margin_of_error = t_critical * std_error

    # Print results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Standard Error: {std_error:.4f}")
    print(f"Margin of Error ({int(confidence*100)}% confidence): {margin_of_error:.4f}")

    return margin_of_error

# Step 2: Use sample data collected by Shweta
sample_data = np.array([72, 75, 70, 74, 73, 76, 75, 74, 72, 73])

# Step 3: Call the function
calculate_margin_of_error(sample_data)

# Interpretation:
# The margin of error tells us how much the sample mean can differ from the true population mean.
# A smaller margin means more precise estimates.
```

```python
# Question:11--->> Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process?

# Answer:11-->>>

# Step 1: Define Bayes' Theorem as a function
def bayes_theorem(prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A):
    # Step 2: Calculate total probability of B
    prob_B = (prior_A * likelihood_B_given_A) + (prior_not_A * likelihood_B_given_not_A)

    # Step 3: Apply Bayes' Theorem to find posterior P(A|B)
    posterior_A_given_B = (prior_A * likelihood_B_given_A) / prob_B

    # Step 4: Print results
    print(f"Posterior Probability P(A|B): {posterior_A_given_B:.4f}")
    return posterior_A_given_B

# Step 5: Use example
# Let's say Sameer tested positive for a disease.
# Prior probability of disease (P(A)) = 0.01
# Probability of positive test if diseased (P(B|A)) = 0.99
# Prior probability of not having disease (P(not A)) = 0.99
# Probability of positive test if not diseased (P(B|not A)) = 0.05

prior_A = 0.01
likelihood_B_given_A = 0.99
prior_not_A = 1 - prior_A
likelihood_B_given_not_A = 0.05

# Step 6: Call the function with values
bayes_theorem(prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A)

# Explanation:
# Bayes' Theorem helps update probability of having a disease given a test result.
# Even with a positive test, the chance of truly having the disease can be low if the disease is rare.
# This is how Bayesian inference works — by updating belief based on new evidence.
```

```python
# Question:12--->> Perform a Chi-square test for independence between two categorical variables in Python?

# Answer:12-->>>

# Import necessary libraries
import numpy as np                           # For data
from scipy.stats import chi2_contingency     # For Chi-square test
import pandas as pd                          # For easier table formatting

# Step 1: Create a contingency table
# Let's say Sameer collected data on students' preference for study method vs. performance
# Rows = Study Method (Online, Offline)
# Columns = Performance (Pass, Fail)
data = np.array([[30, 10],   # Online: 30 pass, 10 fail
                 [20, 20]])  # Offline: 20 pass, 20 fail

# Optional: Convert to DataFrame for display
df = pd.DataFrame(data, columns=["Pass", "Fail"], index=["Online", "Offline"])
print("Contingency Table:")
print(df)

# Step 2: Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(data)

# Step 3: Print the results
print(f"\nChi-square Statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

# Step 4: Interpret the result
alpha = 0.05
if p < alpha:
    print("\nReject the null hypothesis: There is a significant relationship between study method and performance.")
else:
    print("\nFail to reject the null hypothesis: No significant relationship between study method and performance.")

# Explanation:
# The Chi-square test checks if there's an association between two categorical variables.
# In this case, it checks if study method (online/offline) affects performance (pass/fail).
```

```python
# Question:14--->> Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution?

# Answer:14-->>>

# Import necessary libraries
import numpy as np                          # For calculations
from scipy.stats import chisquare           # For goodness-of-fit test

# Step 1: Define observed data
# Let's say Sameer observed the number of customers visiting a shop each weekday
observed = np.array([50, 55, 60, 65, 70])  # Mon to Fri

# Step 2: Define expected data
# Expected equal distribution: same number of visitors each day
expected = np.array([60, 60, 60, 60, 60])  # Equal visitors per day assumption

# Step 3: Perform the Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Step 4: Print the result
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 5: Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: Observed data fits the expected distribution.")

# Explanation:
# This test checks whether the actual number of customers is significantly different
# from the expected uniform distribution across the weekdays.
```
